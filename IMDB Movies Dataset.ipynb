## Data Cleaning

```python
data_copy = data.copy()
```

```python
data_copy['Gross'] = data_copy['Gross'].str.replace(',', '').astype('float')
data_copy['Runtime'] = data_copy['Runtime'].apply(lambda text: text.split()[0]).astype('int')
```

```python
data_copy.head(1)
```

```python
type(data_copy.loc[0,'Gross'])
```

```python
type(data_copy.loc[0,'Runtime'])
```

```python
data_copy['Certificate'] = data_copy['Certificate'].replace(np.nan, 'Unknown')
data_copy['Meta_score'] = data_copy['Meta_score'].replace(np.nan, np.mean(data_copy['Meta_score']))
data_copy['Gross'] = data_copy['Gross'].replace(np.nan, np.mean(data_copy['Gross']))
```

## Exploratory Data Analysis

### Top movies by gross

```python
highest_grossing_movie = data_copy.copy()
highest_grossing_movie = highest_grossing_movie.sort_values(by = 'Gross', ascending = False)
highest_grossing_movie.head(10)[['Director','Series_Title','Gross']]
```

### Movie released in respective years

```python
movie_released_by_year = data_copy['Released_Year'].value_counts()
movie_released_by_year
```

```python
#function to display name of the movies released in a particular year order by Gross
def print_movie(year):
    y = data_copy.loc[data_copy['Released_Year'] == year]
    y = y.sort_values(by = 'Gross', ascending = False)
    return y[['Director','Series_Title', 'Released_Year','Gross']]
```

```python
print_movie('2015')
```

### Analysis of the gross of a movie vs directors

```python
#top 10 directors with highest gross
top_directors = data_copy.sort_values(by = 'Gross', ascending = False)
top_directors.head(10)[['Director','Series_Title','Released_Year', 'Gross']]
```

```python
top_directors['Director'].head(10)
```

```python
#function to display gross of each movie for each director
def print_director(director_name):
    temp = data_copy.loc[data_copy['Director'] == director_name]
    temp = temp.sort_values(by = 'Gross', ascending = False)
    return temp[['Director', 'Series_Title', 'Released_Year','Genre','Gross']]
```

```python
print_director('Christopher Nolan')
```

### Analysis of the No of votes of a movie vs directors

```python
#Top voted movies 
votes = data_copy.sort_values(by = 'No_of_Votes', ascending = False)
votes.head(10)[['Director', 'Series_Title', 'No_of_Votes']]
```

```python
#function to display the no of votes along with IMDB rating of all the movie for a particular director
def print_NoOfVotes(director_name):
    temp1 = data_copy.loc[data_copy['Director'] == director_name]
    temp1 = temp1.sort_values(by = 'No_of_Votes', ascending = False)
    return temp1[['Director', 'Series_Title','IMDB_Rating', 'No_of_Votes']]
```

```python
print_NoOfVotes('Peter Jackson')
```

### Stars with most occurences in movies

```python
temp1 = data_copy['Star1'].value_counts()
temp2 = data_copy['Star2'].value_counts()
temp3 = data_copy['Star3'].value_counts()
temp4 = data_copy['Star4'].value_counts()
```

```python
list1 = [temp1,temp2,temp3,temp4]
df = pd.DataFrame(list1)
df.loc[:,"Tom Hanks":"Aamir Khan"]
```

```python
df.loc['total'] = df.sum()
df
```

```python
#Total number of movies for top 10 actors 
df_transpose = df.transpose()
df_transpose = df_transpose.sort_values(by = 'total', ascending = False)
df_transpose.head(10)['total']
```
